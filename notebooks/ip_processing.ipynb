```python
import sounddevice as sd
import soundfile as sf

# Parameters
duration = 15  # seconds to record
fs = 44100    # sample rate (CD quality)

print("Recording... Speak or hum now!")
# Record audio
recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
sd.wait()  # Wait until recording is finished
print("Recording complete!")

# Save to file
audio_path = 'D:/shree/Miniproject/voicetomusic/data/voice_sample.wav'
sf.write(audio_path, recording, fs)
print(f"Audio saved at {audio_path}")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the recorded file
y, sr = librosa.load(audio_path, sr=None)  # sr=None preserves original sampling rate

# Plot waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Voice Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np

# Extract pitch using librosa.pyin
f0, voiced_flag, voiced_probs = librosa.pyin(
    y,
    fmin=librosa.note_to_hz('C2'),
    fmax=librosa.note_to_hz('C7')
)

# Convert Hz to MIDI notes
midi_notes = librosa.hz_to_midi(f0)

# Plot detected MIDI notes
plt.figure(figsize=(12, 4))
plt.plot(midi_notes, label='Detected MIDI Notes', marker='o', markersize=2)
plt.title("Pitch over Time (MIDI Notes)")
plt.xlabel("Frames")
plt.ylabel("MIDI Note")
plt.legend()
plt.show()
```

```python
import pretty_midi

# Create PrettyMIDI object
pm = pretty_midi.PrettyMIDI()
instrument = pretty_midi.Instrument(program=0)  # Acoustic Grand Piano

# Approximate duration per frame
frame_duration = librosa.get_duration(y=y, sr=sr) / len(midi_notes)

# Convert detected MIDI notes to PrettyMIDI notes
for i, note in enumerate(midi_notes):
    if np.isnan(note):
        continue  # skip silent frames
    midi_note = int(note)
    start = i * frame_duration
    end = start + frame_duration
    pm_note = pretty_midi.Note(velocity=100, pitch=midi_note, start=start, end=end)
    instrument.notes.append(pm_note)

pm.instruments.append(instrument)

# Save MIDI file
output_path = r'D:\shree\Miniproject\voicetomusic\outputs\voice_melody.mid'
pm.write(output_path)
print(f"MIDI file saved at {output_path}")
```
