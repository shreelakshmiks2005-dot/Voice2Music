```python
# ===============================
# All-in-One Voice2Music Input Processing
# ===============================
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import sounddevice as sd
import soundfile as sf
import pretty_midi

# -------------------------------
# Setup folders
# -------------------------------
os.makedirs(r'D:\shree\Miniproject\voicetomusic\data', exist_ok=True)

def record_audio(filename='sample1.wav', duration=15, fs=44100):
    """
    Records audio from microphone and saves as WAV.
    duration: recording time in seconds
    filename: name of the file to save
    """
    print(f"Recording for {duration} seconds...")
    recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
    sd.wait()
    
    audio_path = os.path.join(r'D:\shree\Miniproject\voicetomusic\data', filename)
    sf.write(audio_path, recording, fs)
    
    print(f"Audio saved at {audio_path}")
    return audio_path

# Call function with 15 seconds
audio_path = record_audio(filename='sample1.wav', duration=15)

# -------------------------------
# Load audio & extract pitch
# -------------------------------
y, sr = librosa.load(audio_path, sr=None)
f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
midi_notes = librosa.hz_to_midi(f0)

# Plot pitch
plt.figure(figsize=(12, 4))
plt.plot(midi_notes, label='Detected MIDI Notes', marker='o', markersize=2)
plt.title(f"Pitch over Time (MIDI Notes) - {os.path.basename(audio_path)}")
plt.xlabel("Frames")
plt.ylabel("MIDI Note")
plt.legend()
plt.show()

# -------------------------------
# Save MIDI
# -------------------------------
pm = pretty_midi.PrettyMIDI()
instrument = pretty_midi.Instrument(program=0)  # Acoustic Grand Piano
frame_duration = librosa.get_duration(y=y, sr=sr) / len(midi_notes)

for i, note in enumerate(midi_notes):
    if np.isnan(note):
        continue
    midi_note = int(note)
    start = i * frame_duration
    end = start + frame_duration
    pm_note = pretty_midi.Note(velocity=100, pitch=midi_note, start=start, end=end)
    instrument.notes.append(pm_note)

pm.instruments.append(instrument)
output_name = os.path.basename(audio_path).replace('.wav', '.mid')
output_path = r'D:\shree\Miniproject\voicetomusic\outputs\{output_name}'
pm.write(output_path)
print(f"MIDI saved at {output_path}")
```
